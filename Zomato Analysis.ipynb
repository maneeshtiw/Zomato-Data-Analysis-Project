```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Drop column which is not required for analysis.
data = data.drop(['url','address','phone','menu_item','dish_liked','reviews_list'], axis =1)
```

### Dropping Duplicates

```python
# drop duplicate value present in dataset.
data.drop_duplicates(inplace = True)
```

```python
data.shape
```

### Cleaning Rate Column

```python
# Check unique value in [rate] column
data['rate'].unique()
```

### Removing "NEW" , "-" and "/5" from Rate Column

```python
data["rate"]
```

```python
#Replacing NEW and nan with np.nan
data["rate"] = data["rate"].replace("NEW", np.NaN)
data["rate"] = data["rate"].replace("nan", np.NaN)
data["rate"] = data["rate"].replace("-", np.NaN)
```

```python
# apply lambda function for split data
data["rate"] = data["rate"].apply(lambda x : str(x).split("/")[0])
```

```python
data["rate"] = data["rate"].str.strip()
```

```python
data["rate"].unique()
```

```python
data["rate"] = data["rate"].astype("float")
```

```python
# null value in [rate] column
data['rate'].isnull().sum()
```

### Filling Null Values in Rate Column with Mean

```python
#fill mean value where null data in [rate] column.
data['rate'].fillna(data['rate'].mean(), inplace = True)
```

```python
data['rate'].isnull().sum()
```

```python
data.info()
```

### Dropping Null Values

```python
# drop all null value in the dataset
data.dropna(inplace = True)
```

```python
data.info()
```

```python
#Rename column name [approx_cost(for two people)] and [ listed_in(type) ] and [listed_in(city)]:
data.rename(columns = {'approx_cost(for two people)':'Cost2plates','listed_in(type)':'Type','listed_in(city)':'City'},inplace = True)
```

```python
data.head()
```

```python
# [location] column unique value
data['location'].unique()
```

### Listed in(city) and location, both are there, lets keep only one.

```python
# Here i find both column are same then drop [city] column.
data['City'].unique()
```

```python
# drop [city] column
data.drop('City',axis =1)
```

```python
data.head(2)
```

### Removing , from Cost2Plates Column

```python
# In this data few value are contain (,) it's value is string then python not understand it is integer or float then ---
#  replace every (,)  where present.
data['Cost2plates'].unique()
```

```python
# replace (,) 
data["Cost2plates"] = data["Cost2plates"].str.replace(",", "")
```

```python
data["Cost2plates"].unique()
```

```python
data.head(2)
```

### Cleaning Rest Type Column

```python
# value count in [rest_type]
data['rest_type'].value_counts()
```

```python
rest_types = data['rest_type'].value_counts(ascending = False)
rest_types
```

### Making Rest Types less than 1000 in frequency as others

```python
# divide in multiple part where value is less then 10000
rest_types1000 = rest_types[rest_types < 1000]
```

```python
# create function
def handel_rest_type(value):
    if (value in rest_types1000):
        return "others"
    else:
        return value
```

```python
#Apply fun
data["rest_type"] = data["rest_type"].apply(handel_rest_type)
data["rest_type"].value_counts()
```

### Cleaning Location Column

```python
# clean location
data['location'].value_counts()
```

```python
location = data['location'].value_counts(ascending = False)
location
```

```python
location300 = location[location < 300]
```

```python
def handel_locations(value):
    if (value in location300):
        return "others"
    else:
        return value
```

```python
data["location"] = data["location"].apply(handel_locations)
data["location"].value_counts()
```

```python
#clean cuisines column
data.head(1)
```

### Cleaning Cuisines Column

```python
data['cuisines'].value_counts()
```

```python
cuisines = data['cuisines'].value_counts(ascending = False)
cuisines
```

```python
cuisines100 = cuisines[cuisines < 100]
```

```python
def handel_cuisines(value):
    if (value in cuisines100):
        return "others"
    else:
        return value
```

```python
data["cuisines"] = data["cuisines"].apply(handel_cuisines)
data["cuisines"].value_counts()
```

#### Count Plot of Various Locations

```python
# visualization
plt.figure(figsize = (16,10))
ax = sns.countplot(data['location'])
plt.xticks(rotation=90)
```

```python
data1.plot(kind = 'bar', figsize = (15,8))
```
